# Electrochemistry Notebooks

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from ipywidgets import interactive
```

# Lets see the effect of transfer coefficient on current density

```python
#Eta and f as domain for Butler-Volmer Equation
fn = np.linspace(-4,4,100)

#Butler-Volmer Equation as a function
def plotter(a):
    jac = np.exp((1-a)*fn)-np.exp(-a*fn)
    plt.figure(figsize=(10,6),facecolor='White')
    plt.grid()
    plt.plot(fn,jac, linewidth=3.0,color="Green",label="Net current density")
    plt.legend(prop={'family':'Arial', 'size':12})
    plt.xlabel("Overpotential, $nf$",fontname="Candara",fontsize=18)
    plt.ylabel("Current Density ($mA$ $cm^{-2}}$)",fontname="Candara", fontsize=18)
    

#interactive plot
iplot = interactive(plotter,
                   a=(0,1,0.1))
iplot
```

# Comparing Anodic and Cathodic Current Densities

```python
#Eta and f as domain for Butler-Volmer Equation
fn = np.linspace(-4,4,100)

#Butler-Volmer Equation as a function
def plotter(a):
    ja = np.exp((1-a)*fn)
    jc = np.exp(-a*fn)
    plt.figure(figsize=(10,6),facecolor='White')
    plt.grid()
    plt.plot(fn,ja,linewidth=3.0,label="Anodic current density")
    plt.plot(fn,jc, color="Red",linewidth=3.0, label="Cathodic current density")
    plt.legend(prop={'family':'Arial', 'size':12})
    plt.xlabel("Overpotential, $nf$",fontname="Candara",fontsize=18)
    plt.ylabel("Current Density ($mA$ $cm^{-2}}$)",fontname="Candara", fontsize=18)
    

#interactive plot
iplot = interactive(plotter,
                   a=(0,1,0.1))
iplot
```

# Combining all in one plot

```python
#Eta and f as domain for Butler-Volmer Equation
fn = np.linspace(-4,4,100)

#Defining a function to be called as interactive plot using ipywideget
def plotter(a):
    jac = np.exp((1-a)*fn)-np.exp(-a*fn) #Butler-Volmer equation
    ja = np.exp((1-a)*fn) #Anodic current density only
    jc =-np.exp(-a*fn) #Cathodic current density only
    plt.figure(figsize=(10,6),facecolor='White')
    plt.grid()
    plt.plot(fn,jac, linewidth=2.5,color="Green",label="Net current density")
    plt.plot(fn,ja, '--',linewidth=2.5,color="Blue",label="Anodic current density")
    plt.plot(fn,jc,'--',linewidth=2.5,color="Red",label="Cathodic current density")
    plt.legend(prop={'family':'Arial', 'size':12})
    plt.xlabel("Overpotential, $nf$",fontname="Arial",fontsize=18)
    plt.ylabel("Current Density ($mA$ $cm^{-2}}$)",fontname="Arial", fontsize=18)
    

#interactive plot
iplot = interactive(plotter,
                   a=(0,1,0.1))
iplot
```

```python
#Overpotential (eta)
n = np.array([50, 100, 150, 200, 250])

#Log natural of current denisty (j)
ln_j= np.array([1.48, 2.53, 3.37, 4.18, 5.00])

#Linear regression parameters
slope, intercept, r_value, p_value, std_err = stats.linregress(n, ln_j)
#Best fit line
best_fit_line = slope*n+intercept

#Plots
plt.figure(figsize=(10,6),facecolor='White')
plt.scatter(n, ln_j, label='original data')
plt.plot(n,best_fit_line, color='red', label='Best fit')    
plt.xlabel('Overpotential',fontname="Candara",fontsize=18)
plt.ylabel('lnj',fontname="Candara",fontsize=18)
plt.legend()
plt.grid()
```
